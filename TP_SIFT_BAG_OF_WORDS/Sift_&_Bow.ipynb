```python
hb = np.array([-1,-2,-1])
ha = np.array([1,0,-1])

def compute_grad(I):

    Ix = conv_separable(I,hb,ha)
    Iy = conv_separable(I,ha,hb)

    return Ix, Iy

# Example
Ix, Iy = compute_grad(I)
plt.imshow(Ix ,  cmap='gray')
plt.colorbar()
plt.show()
plt.imshow(Iy ,  cmap='gray')
plt.colorbar()
plt.show()
```

```python
def compute_grad_mod_ori(I):
    Ix,Iy = compute_grad(I)
    Gn = np.sqrt(Ix**2 + Iy**2)
    Go = compute_grad_ori(Ix, Iy, Gn)
    return Gn, Go
```

```python
def compute_histogram(g_n, g_o):
    """
    g_n and g_o are 4x4 matrices that contain the norm, and the discretized orientation.
    Warning: g_o can contain -1, those values should be discarded.
    """
    hist = np.zeros((8))

    for i in range(8):
      hist[i] = np.sum(g_n[g_o == i])

    g_o = g_o[g_o != -1]
    return hist
  
```

```python
def compute_sift_region(Gn, Go, mask=None):
    t_min=.5
    t_max=.2
    with_l2 = True

    patch_size = 16
    sift = np.zeros((128)) 

    if mask is not None:
        Gn = Gn * mask
    
    idx = 0
    # découpage en régions du patch 4*4 du patch 16*16
    for k in range(0, patch_size, 4):
        for l in range(0, patch_size, 4):
            # Pour chaque région calcul de l'histogramme de gradient
            hist = compute_histogram(Gn[l:l+4,k:k+4], Go[l:l+4,k:k+4])            
            #Concaténation des histogrammes des régions pour le SIFT du patch
            sift[idx:idx+8] = hist
            idx += 8


    # Post processing sur Penc
    norm = np.linalg.norm(sift, ord=2)
    # min thresholding on norm
    if norm <= t_min:
        return np.zeros((128))
    # l2-normalization
    if with_l2:
        sift = sift / norm
    # max thresholding on values
    sift[sift >= t_max] = t_max
    # l2-normalization
    if with_l2:
        norm = np.linalg.norm(sift, ord=2)
        sift = sift / norm
    return sift
```

```python
def compute_sift_image(I):
    x, y = dense_sampling(I)
    im = auto_padding(I)
    m = gaussian_mask()
    

    # compute on the global image (norm, gradients)
    Gn, Go = compute_grad_mod_ori(im)

    sifts = np.zeros((len(x), len(y), 128))
    for i, xi in enumerate(x):
        for j, yj in enumerate(y):

            sifts[i, j, :] = compute_sift_region(Gn[xi:xi+16,yj:yj+16],Go[xi:xi+16,yj:yj+16],mask = m) # SIFT of the patch of coords (xi, yj)

    return sifts
```

```python
# Code here the `compute_visual_dict` function
def compute_visual_dict(sift, n_clusters=1000, n_init=1, verbose=1):
    # reorder data
    dim_sift = sift[0].shape[-1]
    sift = [s.reshape(-1, dim_sift) for s in sift]
    sift = np.concatenate(sift, axis=0)
    # remove zero vectors
    keep = ~np.all(sift==0, axis=1)
    sift = sift[keep]
    # randomly pick sift
    ids, _ = compute_split(sift.shape[0], pc=0.05)
    sift = sift[ids]

    # TODO compute kmeans on `sift`, get cluster centers, add zeros vector, and return visual dict
    kmeans = KMeans(n_clusters)
    kmeans.fit(sift)
    vdict =  kmeans.cluster_centers_
    vdict = np.concatenate((vdict, np.zeros((1,128))), axis=0)
    return vdict
```
